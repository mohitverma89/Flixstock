```python
import threading
import time
import math
def first_thread(k=6):    
    startTime = time.time()
    for i in range(0,k):
        endTime = time.time()
        elapsedTime = endTime - startTime
        if (math.trunc(elapsedTime))<21:
            print("Thread 1 is running at %s \n" % math.trunc(elapsedTime))
        else:
            y = threading.Thread(target=second_thread, args=())
            y.start()
            y.join()
        time.sleep(5)


def second_thread():
    startTime = time.time()
    for i in range(0,8):
        endTime = time.time()
        elapsedTime = endTime - startTime
        print("Thread 2 is running at %s \n" % math.trunc(elapsedTime))
        time.sleep(5)


def three_thread():
    startTime = time.time()
    for i in range(0,9):
        endTime = time.time()
        elapsedTime = endTime - startTime
        if (math.trunc(elapsedTime))<39:
            print("Thread 3 is running at %s \n" % math.trunc(elapsedTime))
        else:
            x = threading.Thread(target=first_thread, args=(5,))
            x.start()
        time.sleep(5)

x = threading.Thread(target=first_thread, args=())
x.start()
y = threading.Thread(target=second_thread, args=())
z = threading.Thread(target=three_thread, args=())
z.start()

x.join()
z.join()

```
